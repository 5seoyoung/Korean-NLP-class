```python
from konlpy.tag import Okt
```

```python
file_path = "이인직-혈의_누.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()
```

```python
print(text[:500])
```

(1) 형태소 단위로 분석 (morphs)

```python
from konlpy.tag import Okt

# KoNLPy 객체 생성
okt = Okt()

# 형태소 분석
morphs = okt.morphs(text)
print(morphs[:50])  # 앞 50개 형태소만 출력
```

(2) 품사 태깅 (pos)

```python
# 품사 태깅
pos_tags = okt.pos(text)
print(pos_tags[:50])  # 앞 50개 태깅된 단어 출력
```

(3) 명사만 추출 (nouns)

```python
# 명사 추출
nouns = okt.nouns(text)
print(nouns[:50])  # 앞 50개 명사 출력
```

```python
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 명사 리스트에서 빈도 계산
word_count = Counter(nouns)

# 가장 많이 등장한 상위 30개 단어 출력
print(word_count.most_common(30))

# 워드클라우드 시각화
wordcloud = WordCloud(
    font_path="/usr/share/fonts/truetype/nanum/NanumGothic.ttf",  # 한글 폰트 경로
    background_color="white",
    width=800,
    height=600
).generate_from_frequencies(word_count)

# 그래프 출력
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
