(1) 형태소 단위로 분석 (morphs)

```python
from konlpy.tag import Okt

# KoNLPy 객체 생성
okt = Okt()

# 형태소 분석
morphs = okt.morphs(text)
print(morphs)  # 앞 50개 형태소만 출력
```

(2) 품사 태깅 (pos)

```python
# 품사 태깅
pos_tags = okt.pos(text)
print(pos_tags[:50])  # 앞 50개 태깅된 단어 출력
```

(3) 명사만 추출 (nouns)

```python
# 명사 추출
nouns = okt.nouns(text)
print(nouns)  # 앞 50개 명사 출력
```

```python
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 명사 리스트에서 빈도 계산
word_count = Counter(nouns)

# 가장 많이 등장한 상위 30개 단어 출력
print(word_count.most_common(50))

# 워드클라우드 시각화
wordcloud = WordCloud(
    font_path="/usr/share/fonts/truetype/nanum/NanumGothic.ttf",  # 한글 폰트 경로
    background_color="white",
    width=800,
    height=600
).generate_from_frequencies(word_count)

# 그래프 출력
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. 형태소 분석기 준비
okt = Okt()

file_path = "봄봄.txt"
with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()


# 3. 명사만 추출한 후 문자열로 변환
nouns = okt.nouns(text)
nouns_text = ' '.join(nouns)

# 4. TF-IDF 벡터라이저 적용
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([nouns_text])

# 5. 단어와 점수 추출
words = vectorizer.get_feature_names_out()
scores = tfidf_matrix.toarray()[0]

# 6. 상위 키워드 출력
top_n = 30
df = pd.DataFrame({'word': words, 'score': scores})
top_keywords = df.sort_values(by='score', ascending=False).head(top_n)
print(top_keywords)
```
