```python
from konlpy.tag import Okt
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
text = re.sub(r"[^가-힣\s]", " ", text)
text = re.sub(r"\s+", " ", text).strip()

# 2. 형태소 분석기 호출
okt = Okt()

# 3. 전체 품사 태깅
pos_result = okt.pos(text, norm=True, stem=True)

# 4. 품사 개수 집계
pos_list = [pos for word, pos in pos_result]
pos_counter = Counter(pos_list)

print("전체 품사 분포: ", pos_counter)

# 5. 명사만 추출
nouns = okt.nouns(text)

# 6. 불용어 처리
stopwords = ['그리고', '그런데', '하지만', '또한', '즉', '때문에', '그', '이', '저', '것', '수', '좀']
clean_nouns = [word for word in nouns if word not in stopwords and len(word) > 1]

# 7. 명사 빈도 Top-N 추출
noun_counter = Counter(clean_nouns)
top_nouns = noun_counter.most_common(30)

print("명사 Top-30 빈도수: ", top_nouns)

# 8. 데이터프레임 변환
df_noun = pd.DataFrame(top_nouns, columns=['단어', '빈도수'])

df_noun
```

감정명사/심리동사 빈도 분석

```python
import pandas as pd
from collections import Counter

# 1. 감정명사 사전 불러오기
df_emo = pd.read_csv("감정 신뢰도.txt", sep='\t')  # 탭으로 구분
emo_nouns = df_emo['단어'].tolist()

# 2. 심리동사 사전 불러오기
with open("심리동사.txt", "r", encoding="utf-8") as f:
    psy_verbs = [line.strip() for line in f.readlines()]

# 전처리된 결과 사용
# 명사 리스트 : nouns
# 전체 형태소 리스트 : morphs

# 3. 감정명사 빈도 분석
emo_counter = Counter([word for word in nouns if word in emo_nouns])

# 4. 심리동사 빈도 분석
psy_counter = Counter([word for word in morphs if word in psy_verbs])

# 5. 결과 출력
print("감정명사 Top-20 빈도수:")
print(emo_counter.most_common(20))

print("\n심리동사 Top-20 빈도수:")
print(psy_counter.most_common(20))
```

TF-IDF 벡터화

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. 단어 리스트를 띄어쓰기 기준으로 문자열 변환
doc = ' '.join(nouns)

# 2. TF-IDF 벡터화 객체 생성
tfidf = TfidfVectorizer()

# 3. 벡터화 진행
tfidf_matrix = tfidf.fit_transform([doc])

# 4. 단어별 TF-IDF 값 추출
tfidf_score = dict(zip(tfidf.get_feature_names_out(), tfidf_matrix.toarray()[0]))

# 5. TF-IDF 값 기준 내림차순 정렬
sorted_tfidf = sorted(tfidf_score.items(), key=lambda x: x[1], reverse=True)

# 6. 상위 Top-30 단어 출력
print("TF-IDF Top-30 단어 및 값")
for word, score in sorted_tfidf[:30]:
    print(word, score)

# 7. DataFrame 변환 (시각화용)
df_tfidf = pd.DataFrame(sorted_tfidf, columns=["단어", "TF-IDF 값"])
```

```python
import re
from konlpy.tag import Okt

file_path = "이인직-혈의_누.txt"
with open(file_path, "r", encoding="utf-8") as file:
    raw_text = file.read()  # 정제 안한 원문 그대로 불러오기

# 1. 문장 단위 자르기 (마침표, 느낌표, 물음표 기준)
sentences_origin = re.split('(?<=[.?!])\s+', raw_text)

print("문장 수:", len(sentences_origin))
print("원문 문장 예시:", sentences_origin[:5])

# 2. 형태소 분석 + 명사 추출
okt = Okt()
sentences_nouns = []

for sentence in sentences_origin:
    nouns = okt.nouns(sentence)
    nouns = [word for word in nouns if len(word) > 1]  # 한 글자 제거
    sentences_nouns.append(nouns)

print("문장별 명사 리스트 예시:", sentences_nouns[:5])
```

```python
sentences = []

for idx, nouns in enumerate(sentences_nouns):
    sentences.append(nouns)

# 결과 확인
print(sentences)
```

```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Word2Vec 모델 로드
model = Word2Vec.load("blood_tear_word2vec.model")

# 문장 벡터 생성 함수
def get_w2v_sentence_vector(sentence, model):
    vectors = [model.wv[word] for word in sentence if word in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

# 문장 벡터 전체 생성
sentence_vectors = [get_w2v_sentence_vector(s, model) for s in sentences]

# 특정 문장 기준 유사 문장 검색
target_idx = 5  # 비교할 기준 문장 인덱스
target_vector = sentence_vectors[target_idx]

similarities = cosine_similarity([target_vector], sentence_vectors)[0]

# 유사 문장 Top-5 출력
top_indices = similarities.argsort()[-6:][::-1]  # 자기 자신 포함
print("입력 문장:", sentences[target_idx])
print("유사 문장 Top-5:")
for idx in top_indices[1:]:
    print(f"{sentences[idx]} (유사도: {similarities[idx]})")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# KoBERT 기반 문장 임베딩 모델 로드
model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')

# 문장 임베딩
sentence_vectors = model.encode(sentences)

# 특정 문장 기준 유사 문장 검색
target_idx = 5
target_vector = sentence_vectors[target_idx]

similarities = cosine_similarity([target_vector], sentence_vectors)[0]

# 유사 문장 Top-5 출력
top_indices = similarities.argsort()[-6:][::-1]
print("입력 문장:", sentences[target_idx])
print("유사 문장 Top-5:")
for idx in top_indices[1:]:
    print(f"{sentences[idx]} (유사도: {similarities[idx]})")
```
